```python
#importing all the libraries
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder,StandardScaler,OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
#Reading the csv file
df=pd.read_csv("/content/Crop_recommendation.csv")
df.head()
```

```python
#Here X denotes the input variables for the model and y denotes the target varibale
x = df[['temperature', 'ph', 'rainfall', 'label']]
y = df[['humidity']]
```

```python
x.head()
```

```python
y.head()
```

```python
ct = ColumnTransformer([
    ('encoder', OneHotEncoder(handle_unknown='ignore', sparse_output=False), ['label']),
    ('scaler', StandardScaler(), ['temperature','ph','rainfall'])
])
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
pipeline=Pipeline([
    ('preprocessor', ct),
    ('model', LinearRegression())
])
pipeline.fit(x_train,y_train)
```

```python
y_pred=pipeline.predict(x_test)
```

```python
print("R2 Score:", r2_score(y_test, y_pred))
print("MAE:", mean_absolute_error(y_test, y_pred))
print("MSE:", mean_squared_error(y_test, y_pred))
```

```python
input_data = pd.DataFrame({
    'temperature': [30],
    'ph': [6.6],
    'rainfall': [202.9],
    'label': ['corn']
})

prediction = pipeline.predict(input_data)

print("Predicted Humidity:", prediction[0])
```
